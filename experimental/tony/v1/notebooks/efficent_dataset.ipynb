Process the LMDB file from Seonghwan

```python
from cgflow.data.datasets import EfficentLMDBPocketComplexDataset
```

```python
val_dataset = EfficentLMDBPocketComplexDataset(
    key_path='../../../../experiments/data/complex/zinc_15A/keys/val.txt',
    lmdb_path='../../../../experiments/data/complex/zinc_15A/lmdb/val'
)
```

```python
train_dataset = EfficentLMDBPocketComplexDataset(
    key_path='../../../../experiments/data/complex/zinc_15A/keys/train.txt',
    lmdb_path='../../../../experiments/data/complex/zinc_15A/lmdb/train'
)
```

```python
import cgflow.scriptutil as util
vocab = util.build_vocab()
mol = train_dataset[0].ligand.to_rdkit(vocab)
```

```python
from cgflow.util.rdkit import get_mol_centroid
centroid = get_mol_centroid(mol)
```

```python
holo = train_dataset[0].holo
len(holo)
```

```python
distance_cutoff = 9

subset_holo = holo.select_residues_by_distance(centroid, distance_cutoff)
len(subset_holo)
```

```python
subset_holo
```

```python
subset_holo.write_pdb('test.pdb')
```

```python
from rdkit import Chem

# Write the molecule to a PDB file
Chem.MolToPDBFile(mol, 'test_mol.pdb')

# Append the molecule to the existing test.pdb file
with open('test.pdb', 'a') as f:
    with open('test_mol.pdb', 'r') as f2:
        # Skip the header line
        next(f2)
        # Append the rest of the molecule file
        f.write(f2.read())
        
print("Molecule added to test.pdb")
```
